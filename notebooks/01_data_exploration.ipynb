# 01. Data Exploration Notebook

This notebook conducts Exploratory Data Analysis (EDA) on the raw sensor data of Well B-167 to analyze statistics, correlation structures, and sensor value distributions.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned data
import sys
sys.path.append('../')
from preprocessing.cleaner import load_raw_data, clean_data

raw_path = r'../../DataSet/Data_Export_Selected_B_167_01_Jan_2025_000000_01_Jul_2025_000000 (2).xlsx'
df_raw = load_raw_data(raw_path)
df = clean_data(df_raw)
print("Data loaded. Shape:", df.shape)
```

```python
# Plot distribution of main sensor values
sensor_cols = ['Motor_Current', 'Discharge_Pressure', 'Intake_Pressure', 'Motor_Temperature']
df[sensor_cols].hist(figsize=(10, 8), bins=30)
plt.suptitle('Sensor Distributions')
plt.show()
```

```python
# Plot correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df[sensor_cols].corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Sensor Correlation Matrix')
plt.show()
```
